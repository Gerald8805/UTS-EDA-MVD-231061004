`pecah_billing_address`berfungsi untuk memecah alamat berdasarkan baris baru \n untuk mendapatkan bagian pertama (nomor dan jalan) dan bagian kedua (kota, negara bagian, dan kode pos).

```python
def pecah_billing_address(billing_address):
    parts = billing_address.split('\n')

    if len(parts) == 2:
        nomor_alamat = parts[0].strip()
        kota_dan_kode = parts[1].strip()
        if ',' in kota_dan_kode:
            city, state_postal = kota_dan_kode.split(',')
            state_postal = state_postal.strip().split(' ')
            state = state_postal[0]
            postal_code = state_postal[1] if len(state_postal) > 1 else ''
        else:
            city = kota_dan_kode
            state = ''
            postal_code = ''
    else:
        nomor_alamat, city, state, postal_code = '', '', '', ''

    return pd.Series([nomor_alamat, city.strip(), state, postal_code])

df[['Nomor_Alamat', 'Kota', 'Negara_Bagian', 'Kode_Pos']] = df['billing_address'].apply(pecah_billing_address)

output_file_path = 'updated_hotel_guests_dataset.csv'
df.to_csv(output_file_path, index=False)

print(df[['billing_address', 'Nomor_Alamat', 'Kota', 'Negara_Bagian', 'Kode_Pos']].head())
```

Digunakan untuk menampilkan pie chart dimana sesuai data yang ad dibawah dimana menunjukan berapa persen banyaknya type kamar yang di pesan dimana dari data tersebut menunjukan kalau type kamar yang paling di pesan yaitu kamar type basic dengan presentase sebesar 87,1%

Digunakan untuk menampilkan grafik batang sesuai data yang ada di bawah yang menunjukan bulan paling banyak tamu check in dan yang paing sedikit check in

```python
df['checkin_date'] = pd.to_datetime(df['checkin_date'])
df['Month'] = df['checkin_date'].dt.month
check_in_counts = df['Month'].value_counts().sort_index()

plt.figure(figsize=(10, 6))
plt.bar(check_in_counts.index, check_in_counts.values, color='skyblue')

plt.xlabel('Bulan')
plt.ylabel('Jumlah Tamu Check-in')
plt.title('Jumlah Tamu Check-in per Bulan')
plt.xticks(check_in_counts.index, ['Jan', 'Feb', 'Mar', 'Apr', 'Mei', 'Jun', 'Jul', 'Agu', 'Sep', 'Okt', 'Nov', 'Des'])
plt.show()
```

Menunjukan lama durasi para tamu saat ereka menginap dari awal mereka check in hingga check out, yang dimana tamu tamu yang paling terlama tinggal adalah selama sekitar 83 dari dan yang paling singkat yaitu tamu yabg menginap hanya 1 hari

```python
df['checkin_date'] = pd.to_datetime(df['checkin_date'], errors='coerce', format='%d %b %Y')
df['checkout_date'] = pd.to_datetime(df['checkout_date'], errors='coerce', format='%d %b %Y')

df['stay_duration'] = (df['checkout_date'] - df['checkin_date']).dt.days

valid_stays = df.dropna(subset=['stay_duration'])
valid_stays = valid_stays[valid_stays['stay_duration'] > 0]

sorted_stays = valid_stays.sort_values(by='stay_duration', ascending=False)
top_10_longest_stays = sorted_stays.head(10)
top_10_shortest_stays = sorted_stays.tail(20)

fig, axes = plt.subplots(2, 1, figsize=(10, 10))

axes[0].barh(top_10_longest_stays['guest_email'], top_10_longest_stays['stay_duration'], color='skyblue')
axes[0].set_title('10 tamu yang stay terlama')
axes[0].set_xlabel('Durasi stay(hari)')
axes[0].invert_yaxis()

axes[1].barh(top_10_shortest_stays['guest_email'], top_10_shortest_stays['stay_duration'], color='salmon')
axes[1].set_title('10 tamu yang stay sebentar')
axes[1].set_xlabel('Durasi stay(hari)')
axes[1].invert_yaxis()

plt.tight_layout()
plt.show()
```
